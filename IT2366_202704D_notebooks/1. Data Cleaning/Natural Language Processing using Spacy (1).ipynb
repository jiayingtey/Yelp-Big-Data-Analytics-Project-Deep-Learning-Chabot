### 1.1 Microtext Normalization

#### Microtext is characterized by relaxed spelling and reliance on abbreviations, acronyms, and emoticons
Source: Micro-text Dictionary from https://github.com/SenticNet/microtext-normalization

#### 1.1.1 Removing Emoticons

```
#Removing emoticons
def remove_emoji(string):
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               u"\U00002500-\U00002BEF"  # chinese char
                               u"\U00002702-\U000027B0"
                               u"\U00002702-\U000027B0"
                               u"\U000024C2-\U0001F251"
                               u"\U0001f926-\U0001f937"
                               u"\U00010000-\U0010ffff"
                               u"\u2640-\u2642"
                               u"\u2600-\u2B55"
                               u"\u200d"
                               u"\u23cf"
                               u"\u23e9"
                               u"\u231a"
                               u"\ufe0f"  # dingbats
                               u"\u3030"
                               "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', string)

tips_pdf['text_syntatics_cleaned'] = tips_pdf['text_basic_cleaned'].apply(remove_emoji)
reviews_pdf['text_syntatics_cleaned'] = reviews_pdf['text_basic_cleaned'].apply(remove_emoji)
```

```
# Lematise posts

from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
reviews_pdf['text_syntatics_cleaned'] = [' '.join([lemmatizer.lemmatize(word) for word in reviews.split(' ')])for reviews in reviews_pdf['text_syntatics_cleaned']]
tips_pdf['text_syntatics_cleaned'] = [' '.join([lemmatizer.lemmatize(word) for word in tips.split(' ')])for tips in tips_pdf['text_syntatics_cleaned']]
```

```
import nltk
nltk.download('stopwords')
```

```
#remove stopwords
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
 
stop_words = set(stopwords.words('english'))

def remove_Stopwords(text):
    word_tokens = word_tokenize(text)
    filtered_sentence = [w for w in word_tokens if not w.lower() in stop_words]
    filtered_sentence = []
    for w in word_tokens:
        if w not in stop_words:
            filtered_sentence.append(w)
    return filtered_sentence
 
tips_pdf['text_syntatics_cleaned'] = tips_pdf['text_syntatics_cleaned'].apply(remove_Stopwords)
reviews_pdf['text_syntatics_cleaned'] = reviews_pdf['text_syntatics_cleaned'].apply(remove_Stopwords)
```

```
def join(t):
    return " ".join(t)
tips_pdf['text_syntatics_cleaned'] = tips_pdf['text_syntatics_cleaned'].apply(join)
reviews_pdf['text_syntatics_cleaned'] = reviews_pdf['text_syntatics_cleaned'].apply(join)
```
